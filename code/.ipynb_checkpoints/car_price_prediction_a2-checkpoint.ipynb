```python
# import libraries
import joblib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 0. Preparing the datasets

```python
df = pd.read_csv('Cars.csv')
df.head()
```

```python
# Check unique values to see the categories
df['fuel'].unique(), df['seller_type'].unique(), df['transmission'].unique(), df['owner'].unique()
```

```python
print("Dataset shape before dropping nan values: ", df.shape)
```

```python
df = df.dropna()
print("Dataset shape before after dropping nan values: ", df.shape)
```

```python
# Define a mapping for categorical 'owner' values to numeric values
owner_map = {
    'First Owner': 1,
    'Second Owner': 2,
    'Third Owner': 3,
    'Fourth & Above Owner': 4,
    'Test Drive Car': 5
}

# Apply the mapping to the 'owner' column
# This replaces the categorical strings in 'owner' with their corresponding numeric values
df['owner'] = df['owner'].map(owner_map)

# Display unique values in the 'owner' column after mapping
# This helps verify that the mapping was applied correctly
df['owner'].unique()
```

```python
# For the feature fuel, remove all rows with CNG and LPG 
# because CNG and LPG use a different mileage system i.e., km/kg which is different from kmfeaturepl for Diesel and Petrol
df = df[~df["fuel"].isin(["CNG", "LPG"])]
df.shape
```

```python
# For the feature mileage, remove “kmpl” and convert the column to numerical type (e.g., float).
df.mileage = df.mileage.str.split(' ').map(lambda x: float(x[0]))

# For the feature engine, remove “CC” and convert the column to numerical type (e.g., float)
df.engine = df.engine.str.split(' ').map(lambda x: float(x[0]))

# Do the same for max power
df.max_power = df.max_power.str.split(' ').map(lambda x: float(x[0]))

# For the feature brand, take only the first word and remove the rest
df.name = df.name.str.split(' ').map(lambda x: x[0])

df.head()
```

```python
# Drop the feature torque
df = df.drop('torque', axis=1)

# Test Drive Cars are ridiculously expensive. Since we do not want to involve this, we will simply delete all samples related to it.
df = df.drop(df[df['owner'] == 5].index)
```

```python
def outlier_count(col, data = X_train):
    
    # calculate your 25% quatile and 75% quatile
    q75, q25 = np.percentile(data[col], [75, 25])
    
    # calculate your inter quatile
    iqr = q75 - q25
    
    # min_val and max_val
    min_val = q25 - (iqr*1.5)
    max_val = q75 + (iqr*1.5)
    
    # count number of outliers, which are the data that are less than min_val or more than max_val calculated above
    outlier_count = len(np.where((data[col] > max_val) | (data[col] < min_val))[0])
    
    # calculate the percentage of the outliers
    outlier_percent = round(outlier_count/len(data[col])*100, 2)
    
    if(outlier_count > 0):
        print("\n"+15*'-' + col + 15*'-'+"\n")
        print('Number of outliers: {}'.format(outlier_count))
        print('Percent of data that is outlier: {}%'.format(outlier_percent))
```

```python
# Experiment tracking
import mlflow
import os

mlflow.set_tracking_uri("http://127.0.0.1:5000/")
# Set log name
os.environ["LOGNAME"] = "lou"
# Create mlflow experiment
mlflow.set_experiment(experiment_name="lou-linear-comparison")

if not os.path.exists("poly_transformer"):
    os.mkdir("poly_transformer")
```

### Task 1. Implementation

```python
from sklearn.model_selection import KFold
from mlflow.models.signature import infer_signature

class LinearRegression(object):
    
    #in this class, we add cross validation as well for some spicy code....
    kfold = KFold(n_splits=3)
            
    def __init__(self, regularization, lr=0.001, method='batch', num_epochs=500, batch_size=50, 
                 cv=kfold, init_method="zeros", use_momentum=False, momentum=0.9, feature_names=None, degree=None, run_count=0):
        self.lr         = lr
        self.num_epochs = num_epochs
        self.batch_size = batch_size
        self.method     = method
        self.cv         = cv
        self.init_method = init_method
        self.regularization = regularization
        self.use_momentum = use_momentum
        self.momentum = momentum
        self.velocity = None # Momentum velocity term
        self.feature_names = feature_names
        self.run_count = run_count
        self.degree = degree
        if isinstance(self.degree, int):
            self.poly_transformer = PolynomialFeatures(degree=degree, include_bias=False)

    def mse(self, ytrue, ypred):
        ytrue = np.atleast_1d(ytrue)  
        ypred = np.atleast_1d(ypred)
    
        # Clip extreme values to prevent overflow
        diff = np.clip(ypred - ytrue, -1e10, 1e10)
        
        return np.mean(diff ** 2)

    def r2score(self, ytrue, ypred):
        tss = np.sum(np.square(np.clip(ytrue - np.mean(ytrue), -1e10, 1e10)))
        rss = np.sum(np.square(np.clip(ytrue - ypred, -1e10, 1e10)))
    
        return 1 - (rss / tss)

    def _initialize_weights(self, num_features):
        """
        Initializes weights using either 'zeros' or 'xavier' method.
        """
        if self.init_method == "zeros":
            return np.zeros(num_features)
        elif self.init_method == "xavier":
            lower, upper = - (1.0 / np.sqrt(num_features)), (1.0 / np.sqrt(num_features))
            return np.random.uniform(low=lower, high=upper, size=num_features)
        else:
            raise ValueError("Invalid init_method. Choose 'zeros' or 'xavier'.")
    
    def fit(self, X_train, y_train):
            
        #create a list of kfold scores
        self.kfold_scores = list()
        
        #reset val loss and val r2score
        self.val_loss_old = np.infty
        self.val_r2_old = -np.infty

        #kfold.split in the sklearn.....
        #5 splits
        for fold, (train_idx, val_idx) in enumerate(self.cv.split(X_train)):
                
            X_cross_train = X_train[train_idx]
            y_cross_train = y_train[train_idx]
            X_cross_val   = X_train[val_idx]
            y_cross_val   = y_train[val_idx]

            if isinstance(self.degree, int):
                X_cross_train = self.poly_transformer.fit_transform(X_cross_train)
                X_cross_val = self.poly_transformer.transform(X_cross_val)
            
            # Initialize weights based on user preference
            self.theta = self._initialize_weights(X_cross_train.shape[1])
            
            #define X_cross_train as only a subset of the data
            #how big is this subset?  => mini-batch size ==> 50
            
            #one epoch will exhaust the WHOLE training set
            with mlflow.start_run(run_name=f"run-{self.run_count}-fold-{fold}", nested=True):

                params = {"method": self.method, "lr": self.lr, "initialization": self.init_method, "momentum": self.use_momentum, "reg": type(self).__name__}
                mlflow.log_params(params=params)

                if isinstance(self.degree, int):
                    # Save and Log Polynomial Transformer
                    poly_transformer_path = f"poly_transformer/poly_transformer_run_{self.run_count}_fold_{fold}.pkl"
                    joblib.dump(self.poly_transformer, poly_transformer_path)
                    mlflow.log_artifact(poly_transformer_path, artifact_path=f"run_{self.run_count}_fold_{fold}_poly")
                          
                for epoch in range(self.num_epochs):
                
                    #with replacement or no replacement
                    #with replacement means just randomize
                    #with no replacement means 0:50, 51:100, 101:150, ......300:323
                    #shuffle your index
                    perm = np.random.permutation(X_cross_train.shape[0])
                            
                    X_cross_train = X_cross_train[perm]
                    y_cross_train = y_cross_train[perm]
                    
                    if self.method == 'sto':
                        for batch_idx in range(X_cross_train.shape[0]):
                            X_method_train = X_cross_train[batch_idx].reshape(1, -1) #(11,) ==> (1, 11) ==> (m, n)
                            y_method_train = y_cross_train[batch_idx] 
                            train_loss = self._train(X_method_train, y_method_train)
                    elif self.method == 'mini':
                        for batch_idx in range(0, X_cross_train.shape[0], self.batch_size):
                            #batch_idx = 0, 50, 100, 150
                            X_method_train = X_cross_train[batch_idx:batch_idx+self.batch_size, :]
                            y_method_train = y_cross_train[batch_idx:batch_idx+self.batch_size]
                            train_loss = self._train(X_method_train, y_method_train)
                    else:
                        X_method_train = X_cross_train
                        y_method_train = y_cross_train
                        train_loss = self._train(X_method_train, y_method_train)

                    mlflow.log_metric(key="train_loss", value=train_loss, step=epoch)

                    yhat_val = self.predict(X_cross_val)

                    # Compute metrics
                    val_loss_new = self.mse(y_cross_val, yhat_val)
                    val_r2_new = self.r2score(y_cross_val, yhat_val)
                    
                    mlflow.log_metric(key="val_loss", value=val_loss_new, step=epoch)
                    mlflow.log_metric(key="val_r2score", value=val_r2_new, step=epoch)
                    
                    #early stopping
                    if np.allclose(val_loss_new, self.val_loss_old):
                        break
                    self.val_loss_old = val_loss_new

                # Log model for this fold
                input_example = X_cross_train[:5]  # Sample input for inference
                y_pred_example = self.predict(input_example)
                signature = infer_signature(input_example, y_pred_example)
    
                mlflow.sklearn.log_model(self, artifact_path=f"run_{self.run_count}_fold_{fold}_model", signature=signature, input_example=input_example)
            
                self.kfold_scores.append(val_loss_new)
                print(f"Run: {self.run_count}, Fold: {fold} -> MSE: {val_loss_new:.4f}, R²: {val_r2_new:.4f}")
                        
    def _train(self, X, y):
        yhat = self.predict(X)
        m    = X.shape[0]        
        grad = (1/m) * X.T @(yhat - y)

        # Add regularization term only if a penalty is used
        if self.regularization is not None:
            grad += self.regularization.derivation(self.theta)

        # Apply Momentum-based Update if enabled
        if self.use_momentum:
            if self.velocity is None:
                self.velocity = np.zeros_like(self.theta)
            
            self.velocity = self.momentum * self.velocity + (1 - self.momentum) * grad
            self.theta = self.theta - self.lr * self.velocity
        else:
            self.theta = self.theta - self.lr * grad  # Standard Gradient Descent
        
        return self.mse(y, yhat)
    
    def predict(self, X):
        return X @ self.theta  #===>(m, n) @ (n, )
    
    def _coef(self):
        return self.theta[1:]  #remind that theta is (w0, w1, w2, w3, w4.....wn)

    def _bias(self):
        return self.theta[0]

    def _plot_feature_importance(self):
        """
        Plots feature importance based on absolute values of coefficients.
        Handles polynomial features if applied.
        """
        if self.theta is None:
            raise ValueError("Model has not been trained yet. Fit the model before calling this function.")
    
        feature_importance = np.abs(self._coef())  # Take absolute value of coefficients
    
        # Get feature names
        if isinstance(self.degree, int) and hasattr(self, "poly_transformer"):
            feature_names = self.poly_transformer.get_feature_names_out(self.feature_names)
        else:
            feature_names = self.feature_names if self.feature_names else [f"Feature {i}" for i in range(len(feature_importance))]
    
        # Sort by importance
        sorted_indices = np.argsort(feature_importance)[::-1]  # Descending order
        sorted_features = np.array(feature_names)[sorted_indices]
        sorted_importance = feature_importance[sorted_indices]
    
        # Plot
        plt.figure(figsize=(10, 5))
        plt.barh(sorted_features, sorted_importance, color='skyblue')
        plt.xlabel("Coefficient Magnitude (Feature Importance)")
        plt.ylabel("Features")
        plt.title("Feature Importance Based on Coefficients")
        plt.gca().invert_yaxis()  # Highest importance at top
        plt.show()
```

```python
class LassoPenalty:
    def __init__(self, l):
        self.l = l # lambda value
        
    def __call__(self, theta): #__call__ allows us to call class as method
        return self.l * np.sum(np.abs(theta))
        
    def derivation(self, theta):
        return self.l * np.sign(theta)
    
class RidgePenalty:
    def __init__(self, l):
        self.l = l
        
    def __call__(self, theta): #__call__ allows us to call class as method
        return self.l * np.sum(np.square(theta))
        
    def derivation(self, theta):
        return self.l * 2 * theta
    
class ElasticPenalty:
    def __init__(self, l = 0.1, l_ratio = 0.5):
        self.l = l 
        self.l_ratio = l_ratio

    def __call__(self, theta):  #__call__ allows us to call class as method
        l1_contribution = self.l_ratio * self.l * np.sum(np.abs(theta))
        l2_contribution = (1 - self.l_ratio) * self.l * 0.5 * np.sum(np.square(theta))
        return (l1_contribution + l2_contribution)

    def derivation(self, theta):
        l1_derivation = self.l * self.l_ratio * np.sign(theta)
        l2_derivation = self.l * (1 - self.l_ratio) * theta
        return (l1_derivation + l2_derivation)
    
class Lasso(LinearRegression):
    def __init__(self, method, lr, l, init_method="zeros", use_momentum=False, run_count=0):
        self.regularization = LassoPenalty(l)
        super().__init__(self.regularization, lr, method, init_method=init_method, use_momentum=use_momentum, run_count=run_count)

class Ridge(LinearRegression):
    def __init__(self, method, lr, l, init_method="zeros", use_momentum=False, run_count=0):
        self.regularization = RidgePenalty(l)
        super().__init__(self.regularization, lr, method, init_method=init_method, use_momentum=use_momentum, run_count=run_count)

class ElasticNet(LinearRegression):
    def __init__(self, method, lr, l, l_ratio=0.5, init_method="zeros", use_momentum=False, run_count=0):
        self.regularization = ElasticPenalty(l, l_ratio)
        super().__init__(self.regularization, lr, method, init_method=init_method, use_momentum=use_momentum, run_count=run_count)

class PolynomialRegression(LinearRegression):
    def __init__(self, lr=0.001, method='batch', num_epochs=500, batch_size=50, init_method="zeros", 
                 use_momentum=False, momentum=0.9, degree=2, run_count=0):
        super().__init__(regularization=None, lr=lr, method=method, num_epochs=num_epochs, batch_size=batch_size, init_method=init_method, 
                         use_momentum=use_momentum, momentum=momentum, degree=degree, run_count=run_count)
```

### Task 2. Experiment

```python
import mlflow.sklearn
import numpy as np
from itertools import product
from sklearn.preprocessing import PolynomialFeatures

# Define hyperparameters to compare
regularization_methods = {
    "poly": PolynomialRegression, 
    "lasso": Lasso,
    "ridge": Ridge,
    "normal": LinearRegression,
    "elastic": ElasticNet
}
momentum_options = [False, True]
methods = ["batch", "mini", "sto"]
init_methods = ["zeros", "xavier"]
learning_rates = [0.01, 0.001, 0.0001]

run_count = 0
# Run experiments with MLflow tracking
for reg_name, reg_class in regularization_methods.items():
    for use_momentum, method, init_method, lr in product(momentum_options, methods, init_methods, learning_rates):

        # Create a descriptive run name
        run_name = f"{reg_name} | {method} | {init_method} | lr={lr} | momentum={use_momentum}"
        print(run_name)
            
        # Set regularization parameter (lambda)
        l = 0.1  # Common regularization strength
        l_ratio = 0.5  # ElasticNet L1/L2 ratio

        # Instantiate model
        if reg_name == "normal":
            model = LinearRegression(regularization=None, lr=lr, method=method, init_method=init_method, use_momentum=use_momentum, run_count=run_count)
        elif reg_name == "elastic":
            model = ElasticNet(method=method, lr=lr, l=l, l_ratio=l_ratio, init_method=init_method, use_momentum=use_momentum, run_count=run_count)
        elif reg_name == "poly":
            model = PolynomialRegression(lr=lr, method=method, init_method=init_method, use_momentum=use_momentum, run_count=run_count)
        else:
            model = reg_class(method=method, lr=lr, l=l, init_method=init_method, use_momentum=use_momentum, run_count=run_count)

        run_count += 1
        
        model.fit(X_train, y_train)
```

## 5. Report

The MLflow experiment was conducted to evaluate different regression models, including Polynomial Regression, Lasso, Ridge, and Normal Regression. Multiple runs were executed, and the performance metrics were logged for each fold. The experiment tracked validation loss (val_loss), training loss (train_loss), and R² score (val_r2score) to assess model accuracy and generalization.

The best-performing model in terms of validation loss and R² score was Polynomial Regression (run-27-fold-2), which achieved a validation loss of 0.0856 and a validation R² score of 0.8771. This model was trained with stochastic gradient descent (sto), Xavier initialization, and a learning rate of 0.0001. The training progress plots indicate a rapid decline in loss during early epochs, suggesting effective convergence. The R² score stabilizes after the initial training phase, confirming that the model captures the variance in the data well.

Overall, the experiment successfully identified Polynomial Regression as the most effective model under the tested conditions. Future improvements could include fine-tuning hyperparameters or exploring additional regularization techniques to further optimize performance.
